```python
import cv2
import numpy as np
my_picture = cv2.imread('Picture/BqlRG.jpg')
cv2.imshow('MyPicture',my_picture)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(my_picture.shape)
print(my_picture.size)
```

## Изменение размеров изображения

```python
res = cv2.resize(my_picture,
                 None,
                 fx=.5,
                 fy=.5 , 
                 interpolation = cv2.INTER_CUBIC)
cv2.imshow('MyPicture',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
final_wide = 200
r = float(final_wide) / my_picture.shape[1]
dim = (final_wide,int(my_picture.shape[0] * r))

resized = cv2.resize(my_picture,dim,interpolation = cv2.INTER_CUBIC)
cv2.imshow('MyPicture',resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Отрожение


```python
flip_h = cv2.flip(my_picture, -1)
cv2.imshow("Flip", flip_h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Поворот

```python
h,w = my_picture.shape[:2]
M = cv2.getRotationMatrix2D((w/2, h/2), 80, 1.0)
rotated = cv2.warpAffine(my_picture, M, (w, h))
cv2.imshow("Rotate", rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Изменение цветового пространства

```python
gray = cv2.cvtColor(my_picture,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# HSV (Hue, Saturation, Value)

```python
hsv = cv2.cvtColor(my_picture, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV", hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# LAB (Lightness, a, b)

```python
lab = cv2.cvtColor(my_picture, cv2.COLOR_BGR2LAB)
cv2.imshow("LAB", lab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Работа с фильтрами

# Размытие путем усреднения

```python
average_image = cv2.blur(my_picture,(7,7))
cv2.imshow('Blur',average_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Размытие по Гауссу

```python
gaussian_image = cv2.GaussianBlur(my_picture,(7,7),0)
cv2.imshow('GaussianBlur',gaussian_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Median Blur

```python
gray = cv2.cvtColor(my_picture, cv2.COLOR_BGR2GRAY)
median = cv2.medianBlur(gray, 5)
cv2.imshow("MedianBlur", median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Bilateral Filter

```python
bilateral = cv2.bilateralFilter(my_picture, d=9, sigmaColor=75, sigmaSpace=75)
cv2.imshow("BilateralFilter", bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Laplacian

```python
gray = cv2.cvtColor(my_picture, cv2.COLOR_BGR2GRAY)#Работает тока с яркостью
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)#uint8 - переводим
cv2.imshow("Laplacian", laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Выделение контуров

```python
gray = cv2.cvtColor(my_picture,cv2.COLOR_BGR2GRAY)
#Установка итогового значения 
thresh = 100
#Получение изображения с использованием функций 
_,thresh_img = cv2.threshold(gray,thresh,255,cv2.THRESH_BINARY)
#Обнаружние контуров
contours,hierachy = cv2.findContours(thresh_img,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
#Создание пустого изображения 
img_contours = np.zeros(my_picture.shape)
#Отображение контуров
cv2.drawContours(img_contours,contours,-1,(255,255,255),1)
cv2.imshow("img_contours", img_contours)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Преобразование в оттенки серого
gray = cv2.cvtColor(my_picture, cv2.COLOR_BGR2GRAY)
#Выделение границ
edges = cv2.Canny(gray, 80, 160)
#Поиск контуров
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#Выбор наибольшего контура
cnt = max(contours, key=cv2.contourArea)
#Построение выпуклой оболочки
hull = cv2.convexHull(cnt)
#Создание копии изображения
img_hull = my_picture.copy()
#Отображение выпуклой оболочки
cv2.drawContours(img_hull, [hull], -1, (0, 255, 0), 2)
cv2.imshow("Convex Hull", img_hull)
cv2.waitKey(0)
cv2.destroyAllWindows()
#Получаем внешнию форму самого крупного контура 


```

```python
#Преобразование в оттенки серого
gray = cv2.cvtColor(my_picture, cv2.COLOR_BGR2GRAY)
#Выделение границ
edges = cv2.Canny(gray, 80, 160)
#Поиск контуров
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#Выбор наибольшего контура
cnt = sorted(contours, key=cv2.contourArea)
#Построение ограничивающего прямоугольника
x, y, w, h = cv2.boundingRect(cnt[-3])
#Создание копии изображения
img_rect = my_picture.copy()
#Отображение прямоугольника
cv2.rectangle(img_rect, (x, y), (x + w, y + h), (255, 0, 0), 2)
cv2.imshow("Bounding Rectangle", img_rect)
cv2.waitKey(0)
cv2.destroyAllWindows()



```
